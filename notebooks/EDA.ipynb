A maioria das pessoas mortas pela polícia americana são homens.

```python
usPop = pd.read_csv('../datasets/acs2017_census_tract_data.csv')

usPop['Other'] = round(100 - usPop['Hispanic'] - usPop['White'] - usPop['Black'] - usPop['Native'] - usPop['Asian'], 2)
orderedRace = ['Asian','Black','Hispanic','Native','Other','White']

usRacePop = pd.DataFrame(columns = [])
arrayRacePop = []
mortalityPerMillion = []
usTotalPop = usPop['TotalPop'].sum()

for item in orderedRace:
    usRacePop[item + 'Pop'] = round(usPop[item] * usPop['TotalPop'] / 100)
    arrayRacePop.append(round(usRacePop[item+'Pop'].sum() / usTotalPop, 4))

print(arrayRacePop)
#Obs: podem haver pequenas variações nos dados pelo uso do round nos métodos acima
```

```python
#Número de mortes por raça
labels, counts = np.unique(df['race'], return_counts=True)
plt.bar(labels, counts, align='center', edgecolor='k')
plt.gca().set_xticks(labels)
plt.show()
```

Em geral, a polícia americana mata mais brancos, seguidos de negros e hispânicos. Agora, podemos comparar esses números com o tamanho de cada população nos Estados Unidos. 

```python
#Taxa de mortes por raça, por milhão de habitantes

labels, counts = np.unique(df['race'], return_counts=True)

Overall = counts.sum() / (usTotalPop / 1000000)   

newCount = np.zeros(len(counts))
for i in range(len(counts)):
    newCount[i] = counts[i] / ((usTotalPop) * (arrayRacePop[i]) / 1000000.0)

labels = np.append(labels, 'Overall')
newCount = np.append(newCount, Overall)
    
plt.bar(labels, newCount, align='center', edgecolor='k')
plt.gca().set_xticks(labels)
plt.axhline(y = Overall, color='r')
plt.show()
```

Quando ajustamos as mortes pela população, vemos que a taxa de mortalidade para negros e nativos são mais do que o dobro da taxa média da população, seguidos por hispânicos e depois brancos, que estão um pouco abaixo da média.
Um dos fatores que explicam a alta taxa dos nativos é a sua pequena população, isto é, poucas mortes podem aumentar significativamente essa taxa.

```python
#Números absolutos de mortes desarmados
#Taxa de mortes desarmados por milhão (por raça)
df['unarmed'] = (df['armed'] == 'unarmed').astype(int)
unarmedDeaths = []

for i in range(len(orderedRace)):
    unarmedDeaths.append( df[(df['unarmed'] == 1) & (df['race'] == orderedRace[i])]['unarmed'].sum())
    
    
unarmedDeathsPopAdj = []
for i in range(len(unarmedDeaths)):
    unarmedDeathsPopAdj.append( unarmedDeaths[i] / ((usTotalPop) * (arrayRacePop[i]) / 1000000.0))
    
plt.bar(orderedRace, unarmedDeaths, align='center', edgecolor='k')
plt.gca().set_xticks(orderedRace)
plt.show()


unarmedDeathsPopAdj.append(df['unarmed'].sum() / (usTotalPop / 1000000))
                           
newOrderedRace = orderedRace.copy()
newOrderedRace.append('Overall')

plt.bar(newOrderedRace, unarmedDeathsPopAdj, align='center', edgecolor='k')
plt.gca().set_xticks(newOrderedRace)
plt.axhline(y = unarmedDeathsPopAdj[6], color='r')
plt.show()
```

No primeiro grafico podemos ver as mortes das pessoas desarmadas distribuídas em cada uma das raças. Já no segundo, temos estas mortes ajustadas à proporção da população, desta forma, nota-se que ao contrário do que aparenta na primeira imagem, pessoas brnacos morrem mais, vemos no segundo grafico que, proporcionalmente, os negros morrem muito mais do que as demais raças.

Por fim, temos os estados que ocorreram mais mortes. Sendo eles: California (CA), Texas (TX), Florida (FL), Arizona (AZ) e Colorado (CO). Estes estados se encontram geograficamente mais próximos da região sul do país. Coincidencia? Será possível encontrar variáveis que expliquem esssa distribuição ou ela é aleatória?

```python
#Mortes por estados ajustados a população:
death_state = df.groupby('state').size().sort_index()
pop_state = usPop.groupby('State')['TotalPop'].sum().sort_index()
sigla = ['AL','AK','AZ','AR','CA',
        'CO','CT','DE','DC','FL','GA','HI',
        'ID','IL','IN','IA','KS','KY','LA','ME',
        'MD','MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ','NM','NY'
        ,'ND','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VT','VA',
        'WA','WV','WI','WY']
death_state = death_state.reindex(sigla)
pop_state = pop_state.drop(['North Carolina','Puerto Rico'], axis=0)
prop_death_state = death_state / pop_state.values
prop_death_state = prop_death_state.sort_values(ascending=False)
prop_death_state.plot(kind = 'bar')
```
