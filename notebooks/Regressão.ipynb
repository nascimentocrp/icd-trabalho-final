# Regressão Linear
#### É possivel predizer a idade da pessoa a partir dos dados?

```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('../datasets/shootings.csv')
df.head()
#df.groupby('flee').count()
```

```python
data = df.copy()
binary_gender = data['gender'] =='M'
binary_gender = binary_gender.astype(int)
data['gender'] = binary_gender

data['armed'] = (data['armed'] != 'unarmed').astype(int)

data['flee'] = (data['flee'] != 'Not fleeing').astype(int)

data['signs_of_mental_illness'] = data['signs_of_mental_illness'].astype(int)
data= data.drop(['name','date','id','city','body_camera'], axis=1)
data = pd.get_dummies(data)
data
```

```python
X = data.drop('age',axis=1)
y = data['age']
X['intercept'] = 1
c = X.columns.tolist()
print(c)
c.insert(0,c[-1])
X=X.reindex(columns = c[:-1])
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
sc = StandardScaler()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2)
Z_train = sc.fit_transform(X_train)
sc.fit(X_train)
Z_test = sc.transform(X_test)
X_train.shape
```

```python
def significants(l):
    return [ int(l.index(i)) for i in l if i>-0.3 and i<0.3 ]
```

```python
#model = SGDRegressor(max_iter = np.ceil(10**6 / len(Z_train)),tol=1e-5,shuffle=True)
model = Ridge()
model.fit(Z_train,y_train)
y_hat = model.predict(X_test)
err = mean_squared_error(y_test,y_hat)
r2 = r2_score(y_test,y_hat)
print('Erro quadratico:',err,'\nR²:',r2)
coef = model.coef_.tolist()
print(coef.index(min(model.coef_)))
idxs = significants(coef)
idxs
```

```python
X = X.drop(idxs,axis=1)
X
```
