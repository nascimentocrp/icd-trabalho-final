#### É possivel predizer a idade da pessoa a partir dos dados?

```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('../datasets/shootings.csv')
df.head()
df.groupby('flee').count()
```

```python
data = df.copy()
binary_gender = data['gender'] =='M'
binary_gender = binary_gender.astype(int)
data['gender'] = binary_gender

data['armed'] = (data['armed'] != 'unarmed').astype(int)

data['flee'] = (data['flee'] != 'Not fleeing').astype(int)

data['signs_of_mental_illness'] = data['signs_of_mental_illness'].astype(int)
data= data.drop(['name','date','id','city'], axis=1)
data = pd.get_dummies(data)
data
```
