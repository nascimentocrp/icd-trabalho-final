# Regressão Linear
#### É possivel predizer a idade da pessoa a partir dos dados?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/shootings.csv')
df.head()
df.groupby('threat_level').count()
```

```python
data = df.copy()

data['gender'] = (data['gender'] == 'M').astype(int)
data['armed'] = (data['armed'] != 'unarmed').astype(int)
data['flee'] = (data['flee'] != 'Not fleeing').astype(int)
data['signs_of_mental_illness'] = data['signs_of_mental_illness'].astype(int)

data= data.drop(['name','date','id','city','body_camera','manner_of_death','armed','state',
                 'threat_level'], axis=1)
data = pd.get_dummies(data)
data
```

```python
X = data.drop('age',axis=1)
y = data['age']
X['intercept'] = 1
c = X.columns.tolist()
c.insert(0,c[-1])
X=X.reindex(columns = c[:-1])
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
sc = StandardScaler()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2)
Z_train = sc.fit_transform(X_train)
sc.fit(X_train)
Z_test = sc.transform(X_test)
X_train.shape
```

```python
def significants(l):
    return [ int(l.index(i)) for i in l if i>-0.3 and i<0.3 ]
```

```python
#model = SGDRegressor(max_iter = np.ceil(10**6 / len(Z_train)),tol=1e-5,shuffle=True)
model = Ridge()
model.fit(Z_train,y_train)
y_hat = model.predict(X_test)
err = mean_squared_error(y_test,y_hat)
r2 = r2_score(y_test,y_hat)
print('Erro quadratico:',err,'\nR²:',r2)
```

```python
idxs = significants(model.coef_.tolist())
new_X = X.drop(X.columns[idxs],axis=1)
y1 = df['age']
new_X['intercept'] = 1
c_val = new_X.columns.tolist()
c_val.insert(0,c[-1])
new_X = new_X.reindex(columns=c_val[:-1])
new_X
```

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(new_X,y1,test_size =0.2)
Z1_train = sc.fit_transform(x1_train)
sc.fit(x1_train)
Z1_test = sc.transform(x1_test)

model.fit(Z1_train,y1_train)
y1_hat = model.predict(x1_test)
err = mean_squared_error(y1_test,y1_hat)
r2 = r2_score(y1_test,y1_hat)
print('Erro quadratico:',err,'\nR²:',r2)
```
